## Loading in Dataset

```python
import pandas as pd

data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
data.head()
```

## Preprocessing Data

Transforming the classes to numeric data

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['class'] = le.fit_transform(data['class'])
data.head()
```

Split features and label and transform them to a Numpy Array

```python
import numpy as np

X = np.array(data.drop(['class'], axis=1))
y = np.array(data['class'])
```

## Building Model

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(X, y)
```

```python
help(clf)
```

## Evaluating Accuracy

In this notebook we are going to evaluate the accuracy on the data which we used to train. You shouldn't to this because you can't be sure if the result you get means anything because the model could just overfit the data. In reality we would split the dataset into a training and test set. We will cover this in the next tutorial.

```python
accuracy = clf.score(X, y)
accuracy
```
